```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#from sklearn.model_selection import train_test_split
```

```python
weight_height = pd.read_csv('weight-height.csv')
weight_height.head()
```

```python
weight_height.shape
```

```python
# predictor variable
X  = weight_height[['Height']]
X.head()
```

```python
# target of regression model
y = weight_height[['Weight']]
y.head()
```

### Training and test data split (80-20)


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

### Build the model using ScikitLearn's linear regression estimator object


```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

## Invoke the now trained model on the X_test data


```python
y_pred = regressor.predict(X_test)
```

### Evaluate the model via it's r<sup>2</sup> score

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```

### Reshape to use with matplot lib

```python
#X_train = X_train.reshape(-1,1)
#y_train = y_train.reshape(-1,1)
X_train = X_train.values.reshape(-1,1)
y_train = y_train.values.reshape(-1,1)
```

## Plot the linear regressor with the data

```python
from matplotlib import rcParams

rcParams.update({
    'font.family':'sans-serif',
    'font.sans-serif':['DejaVu Sans'],
})

plt.scatter(X_train, y_train, color = 'y')
plt.plot(X_train, regressor.predict(X_train))
plt.show()
```
